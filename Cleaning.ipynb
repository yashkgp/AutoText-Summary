```python
import xml.etree.ElementTree as ET
from pathlib import Path
errors=[]
for i in range(6,10):
    for j in range(2200):
        file = Path(r'nltk_data/corpora/fulltext/0%d_%d.xml'%(i,j))
        if file.is_file():
            try :
                my_file = open(r'nltk_data/corpora/fulltext/0%d_%d.xml'%(i,j), "r",errors='ignore')
                counter =0
                s=[]
                lines = my_file.readlines()
                for index,line in enumerate(lines) :
                    if(line.find("catchphrase")<=0):
                        s.append(line)
                        counter+=1
                t=' '.join(s)
                u=[]
                root = ET.fromstring(t)
                for sent in root.iter('sentence'):
                    u.append(sent.text)
                print ("%d_%d"%(i,j))
                f= open("nltk_data/corpora/text/%d_%d.txt"%(i,j),"w+")
                for sent in u:
                    f.write(sent+"\n")
                f.flush()
                f.close()
            except SyntaxError :
                print ("added to error 0%d_%d"%(i,j))
                errors.append("0%d_%d"%(i,j))
                my_file = open(r'nltk_data/corpora/fulltext/0%d_%d.xml'%(i,j), "r",errors='ignore')
                counter =0
                s=[]
                lines = my_file.readlines()
                for index,line in enumerate(lines) :
                    if(line.find("catchphrase")<=0):
                        s.append(line)
                        counter+=1
                f= open("nltk_data/corpora/C_file/%d_%d.xml"%(i,j),"w+")
                for sent in s:
                    f.write(sent+"\n")
                f.flush()
                f.close()
                
                
                
            
        


    
    
        
        
```
